```python
def get_abstract(source, keywork):
    
    url = f"https://fr.{source}.org/wiki/{keywork}"
    print(url)
    
    response = requests.get(url)
    print(response)
    
    content_source = response.text.encode("utf-8")
    xml_doc = html.document_fromstring(content_source)
    
    abstract = xml_doc.xpath('//div[@id="toc"]/preceding-sibling::p | //div[@id="toc"]/preceding-sibling::ul | //div[@id="toc"]/preceding-sibling::ol ')
    
    if len(abstract) == 0: # dans wikida il y n'y a parfois pas de table des matières
        abstract = xml_doc.xpath('//h2[1]/preceding-sibling::p | //h2[1]/preceding-sibling::ul | //h2[1]/preceding-sibling::ol ')

    # unidecode.unidecode : https://stackoverflow.com/questions/26068832/how-to-remove-this-xa0-from-a-string-in-python
    # pour enlever les \xa0 notamment
    # toutefois unidecode.unidecode enlève les accents, faire un replace \xa0 si on veut garder les accents
    
    return keywork, ' '.join([a.text_content()for a in abstract]).strip().replace('\n', ' ').replace('\xa0',' ').replace('/r',' ').replace('  ',' ')
```

```python
def insert_all_data(source):
    for keywork in PAGES:
        keywork, abstract = get_abstract(source, keywork)
        output = (source, keywork, abstract) # tuple
        data.append(output) # add tuple to list
        print(f"{keywork} => DONE")
        time.sleep(2)
```
